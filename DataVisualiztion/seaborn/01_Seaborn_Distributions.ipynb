## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data

```python
sns.get_dataset_names()
```

```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

## Disturbutions Plots

### displot

```python
plt.hist(mpg['horsepower'])
```

```python
sns.displot(mpg['horsepower'],kde = True)
```

```python
sns.displot(mpg['horsepower'],kde=False)
```

```python
sns.displot(mpg['horsepower'],kde=False,bins=50)
```

### jointplot

```python
sns.jointplot(x='horsepower',y='mpg',data=mpg,kind='scatter')
```

```python
sns.jointplot(x='horsepower',y='mpg',data=mpg,kind='hex')
```

```python
mpg[['horsepower','mpg']].corr()
```

```python
sns.jointplot(x='horsepower',y='mpg',data=mpg,kind='reg')
```

```python
sns.jointplot(x='horsepower',y='mpg',data=mpg,kind='kde')
```

### pairplot

```python
sns.pairplot(mpg)
```

```python
mpg[['horsepower','weight']].corr()
```

```python
mpg['PowerPerCylinders'] = mpg['horsepower']/mpg['cylinders']
```

```python
mpg[['horsepower','PowerPerCylinders']].corr()
```

```python
mpg.head()
```

```python
sns.pairplot(mpg,hue='cylinders')
```

### rugplot

```python
sns.rugplot(mpg['horsepower'])
```

```python
sns.displot(mpg['horsepower'],kde=False)
```

```python
sns.kdeplot(mpg['horsepower'])
```

### Multi Plots

```python
sns.kdeplot(mpg['horsepower'])
sns.rugplot(mpg['horsepower'])
```
