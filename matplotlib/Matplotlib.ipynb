## Imports

first import necessary library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## lets prepare some data

```python

x = np.linspace(0,5,20)
y = x**2 - 4 * x + 1 
print("x :",x)
print("y :",y)
```

## Regular Matplotlib

### Basics

`.plot` methods used to draw 2d plot 

```python
plt.plot(x,y,'r')
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

### Subplot

to draw two plots in one figure just plot those right after other


```python
plt.plot(x,y,'r')
plt.plot(y,x,'b')
plt.show()
```

to draw two plots in seperate figure, Using `.subplot` like this 


```python

plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'bo-')
plt.show()
```

## OOP Matplotlib

The `figure()` method in pyplot module of matplotlib library is used to create a new figure.

```python
fig = plt.figure()

axes = fig.add_axes([0.8,0.8,0.8,0.8])

axes.plot(x,y,'b')

axes.set_xlabel('Set X Label')
axes.set_ylabel('Set Y Label')
axes.set_title('Set Title')
```

To understand multiple axes( multiple rectangle insertion in generated figure) easily, Think of  a rectangle which is 1 * 1 (with 0.1 as increment ).Within the rectangle we have arrange those axes with specifying ([a,b,c,d])

**(a,b) is the point in southwest corner of the rectangle which we create.**

**c represents width and d represents height of the respective rectangle.**

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

axes1.plot(x,y,'b')
axes1.set_xlabel('Set X Label')
axes1.set_ylabel('Set Y Label')
axes1.set_title('Set Title')

axes2 = fig.add_axes([0.2,0.4,0.4,0.3])
axes2.plot(y,x,'r')
axes2.set_xlabel('Set X Label')
axes2.set_ylabel('Set Y Label')
axes2.set_title('Set Title')
```

```python
fig1 = plt.figure()
fig2 = plt.figure()
axes1 = fig1.add_axes([0.1,0.1,0.8,0.8])

axes1.plot(x,y,'b')
axes1.set_xlabel('Set X Label')
axes1.set_ylabel('Set Y Label')
axes1.set_title('Set Title')

axes2 = fig2.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(y,x,'r')
axes2.set_xlabel('Set X Label')
axes2.set_ylabel('Set Y Label')
axes2.set_title('Set Title')
```

### Subplot

`pyplot.subplots` creates a figure and a grid of subplots with a single call, while providing reasonable control over how the individual plots are created.

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'r')
axes[0].set_xlabel('Set X Label 1')
axes[0].set_ylabel('Set Y Label 1')
axes[0].set_title('Set Title 1')

axes[1].plot(y,x,'b')
axes[1].set_xlabel('Set X Label 2 ')
axes[1].set_ylabel('Set Y Label 2')
axes[1].set_title('Set Title 2')
```

### Combine

another way to create subplot on figure 

```python
fig = plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
fig.savefig('Filename.tiff')
```

### Legends

Place a legend on the Axes.



```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x**2')
ax.plot(x,x**3, label='x**3')
ax.legend()
```

```python
plt.plot(x,x**2, label='x**2')
plt.plot(x,x**3, label='x**3')
plt.legend()
```

```python
plt.plot(x,x**2, label='x**2')
plt.plot(x,x**3, label='x**3')
plt.legend(loc=8)
# 1 --> upper right
# 2 --> upper left
# 3 --> lower left
# 4 --> lower right
# 0 or Nothing --> Let Matplotlib Decide
```

## Plot Range

changing  plot range by setting limit on x , y axis

```python
fig,axes = plt.subplots(1,3,figsize=(12,4))
axes[0].plot(x,x**2,x,x**3)
axes[0].set_title('Default Axes Ranges')

axes[1].plot(x,x**2,x,x**3)
axes[1].axis('tight')
axes[1].set_title('tight axes')

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([0,4])
axes[2].set_title('Custom Axis Range')
```
