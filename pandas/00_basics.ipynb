### Reading DataBase

In pandas, data is stored in DataFrames.  A DataFrame is a two-dimensional data structure composed of rows and columns — exactly like a simple spreadsheet or a SQL table. Each column of a DataFrame is a pandas Series. These columns should be of the same length, but they can be of different data types — float, int, bool, and so on.

Read a comma-separated values (csv) file into DataFrame:

```python
import pandas as pd
df = pd.read_csv('Datasets/USA_Housing.csv')
```

### Check missing value

df.isnull() detects missing values.

```python
df.isnull()
```

It is possible to fill missing values.

```python
# fill missing values with 0
df.fillna(0)

# forward-fill missing values (propagates last valid observation forward to next)
df.fillna(method='ffill')

# backward-fill missing values (propagates next valid observation backward to last)
df.fillna(method='bfill')

# fill missing values using interpolation
df.interpolate()
```

df.dropna() drops rows or columns that have null values.

```python
# Drop row that have null value
df_drop_na = df.dropna()

# drop columns that have null value
df_drop_na = df.dropna(axis=1)

df.info()
```

df.drop() drops rows or columns

```python
# Drop rows 0 and 1
df_drop = df.drop([0,1],axis=0)
df_drop.head()
```

```python
# Drop 'Avg_AreaNumberofRooms' column
df_drop = df.drop(["Avg_AreaNumberofRooms"],axis=1)
df_drop.head()
```

### Pivot Table

pd.pivot_table() Create a spreadsheet-style pivot table as a DataFrame.

```python
# first convert all of float to int
df["Avg_AreaNumberofRooms"] = df["Avg_AreaNumberofRooms"].apply(func=lambda x:int(x))
df["Avg. Area Number of Bedrooms"] = df["Avg. Area Number of Bedrooms"].apply(func=lambda x:int(x))
# Selecting Homes with number of room less than 7
filt_df = df[df['Avg_AreaNumberofRooms']<7]
```

```python

pivot_table = pd.pivot_table(filt_df, index='Avg_AreaNumberofRooms', values='Price', aggfunc='mean')
pivot_table.head()
```

```python

pivot_table = pd.pivot_table(filt_df, index=['Avg_AreaNumberofRooms','Avg. Area Number of Bedrooms'], values='Price', aggfunc='mean')
pivot_table
```

### Group By

Groups DataFrame using a mapper or by a Series of columns. A groupby operation involves some combination of splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and compute operations on these groups.

```python
# first convert all of float to int
df["Avg_AreaNumberofRooms"] = df["Avg_AreaNumberofRooms"].apply(func=lambda x:int(x))
df["Avg. Area Number of Bedrooms"] = df["Avg. Area Number of Bedrooms"].apply(func=lambda x:int(x))
# Selecting Homes with number of room less than 7
filt_df = df[df['Avg_AreaNumberofRooms']<7]
```

```python
grouped = filt_df.groupby('Avg_AreaNumberofRooms')
# Print the mean of each group
grouped.mean(numeric_only=True)
```

```python
grouped = filt_df.groupby(['Avg_AreaNumberofRooms',"Avg. Area Number of Bedrooms"])
# Print the mean of each group
grouped.mean(numeric_only=True)
```

### Merge
df.merge() merges DataFrame or named Series objects with a database-style join. A named Series object is treated as a DataFrame with a single named column. The join is done on columns or indexes. If joining columns on columns, the DataFrame indexes will be ignored. Otherwise if joining indexes on indexes or indexes on a column or columns, the index will be passed on. When performing a cross merge, no column specifications to merge on are allowed.

```python
# Create the first DataFrame
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                   'value': [1, 2, 3, 4]})

# Create the second DataFrame
df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'],
                   'value': [5, 6, 7, 8]})

# Merge the two DataFrames on the 'key' column
merged_df = df1.merge(df2, on='key')

# Print the merged DataFrame
print(merged_df)
```
