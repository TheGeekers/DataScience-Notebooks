# Pandas Built-in Data Visualization

## Data

```python
base = datetime.date(2000,1,1)
dates = [str(base + datetime.timedelta(days=x)) for x in range(1000)]
data_1 = {}
data_1['dates'] = np.array(dates)
data_1['A'] = np.random.rand(1000)
data_1['B'] = np.random.rand(1000)
data_1['C'] = np.random.rand(1000)
data_1['D'] = np.random.rand(1000)
df1 = pd.DataFrame(data=data_1,columns=['A','B','C','D'], index=data_1['dates'])
df1.head()
```

```python
data_2 = {}
data_2['a'] = np.random.rand(10)
data_2['b'] = np.random.rand(10)
data_2['c'] = np.random.rand(10)
data_2['d'] = np.random.rand(10)
df2 = pd.DataFrame(data=data_2,columns=['a','b','c','d'])
df2
```

### Area
An area plot displays quantitative data visually.

```python
df2.plot.area(alpha=0.4)
```

```python
df2.loc[0].sum()
```

### Barplots
A bar plot is a plot that presents categorical data with rectangular bars with lengths proportional to the values that they represent. A bar plot shows comparisons among discrete categories. One axis of the plot shows the specific categories being compared, and the other axis represents a measured value.

```python
df2.plot.bar(figsize=(12,3))
```

horizontal bar plot.

```python
df2.plot.barh(figsize=(3,12))
```

Stacked form.

```python
df2.plot.bar(stacked=True)
```

### Histograms

A histogram is a representation of the distribution of data. This function groups the values of all given Series in the DataFrame into bins and draws all bins.

```python
df1['A'].plot.hist(bins=50)
```

### Line Plots

Plot Series or DataFrame as lines.

```python
df1.reset_index().plot.line(x='index',y='B',figsize=(12,3),lw=1)
```

### Scatter Plots

Create a scatter plot with varying marker point size and color. The coordinates of each point are defined by two dataframe columns and filled circles are used to represent each point. This kind of plot is useful to see complex correlations between two variables.

```python
df1.plot.scatter(x='A',y='B')
```

sns  hue   ----->   pandas c

sns  pallete ---->   cmap

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

sns  hue   ----->   pandas s

sns  pallete ---->   cmap

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*200)
```

### Box Plots
A box plot is a method for graphically depicting groups of numerical data through their quartiles. The box extends from the Q1 to Q3 quartile values of the data, with a line at the median (Q2). The whiskers extend from the edges of box to show the range of the data. The position of the whiskers is set by default to 1.5*IQR (IQR = Q3 - Q1) from the edges of the box. Outlier points are those past the end of the whiskers.

```python
df2.plot.box()
```

### hexbin
hexagonal binning plot.

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```

```python
df.plot.scatter(x='a',y='b',cmap='Oranges')
```

### KDE Plots
Kernel Density Estimate plot using Gaussian kernels. In statistics, kernel density estimation (KDE) is a non-parametric way to estimate the probability density function (PDF) of a random variable. This function uses Gaussian kernels and includes automatic bandwidth determination.

```python
df2['a'].plot.kde()
```

```python
df2.plot.kde()
```

```python
df2.plot.density()
```

### Pie plot
A pie plot is a proportional representation of the numerical data in a column.

```python
df2['a'].plot.pie()
```
