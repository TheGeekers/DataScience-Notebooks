# Pandas Built-in Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Pandas Built-in Data Visualization

## Data

```python
df1 = pd.read_csv('df1')
```

```python
df1
```

```python
df1 = pd.read_csv('df1',index_col=0)
```

```python
df1
```

```python
df2=pd.read_csv('df2')
```

```python
df2
```

### Area

```python
df2.plot.area(alpha=0.4)
```

```python
df2.loc[0].sum()
```

### Barplots

```python
df2.plot.bar(figsize=(12,3))
```

```python
df2.plot.barh(figsize=(3,12))
```

```python
df2.plot.bar(stacked=True)
```

### Histograms

```python
df1.head()
```

```python
df1['A'].plot.hist(bins=50)
```

### Line Plots

```python
df1
```

```python
df1.reset_index().plot.line(x='index',y='B',figsize=(12,3),lw=1)
```

## Scatter Plots

```python
df1.plot.scatter(x='A',y='B')
```

sns  hue   ----->   pandas c

sns  pallete ---->   cmap

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

sns  hue   ----->   pandas s

sns  pallete ---->   cmap

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*200)
```

### Box Plots

```python
df2.plot.box()
```

## hexbin

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```

```python
df.plot.scatter(x='a',y='b',cmap='Oranges')
```

## KDE Plots

```python
df2['a'].plot.kde()
```

```python
df2.plot.kde()
```

```python
df2.plot.density()
```

```python
df2
```

```python
df2['a'].plot.pie()
```

## Style Sheet

```python
df1['A'].hist()
```

```python
plt.style.use('ggplot')
df1['A'].hist()
```

```python
plt.style.use('bmh')
df1['A'].hist()
```

```python
plt.style.use('dark_background')
df1['A'].hist()
```

```python
plt.style.use('fivethirtyeight')
df1['A'].hist()
```
