## Data

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Regression Plots

### lmplot

Plot data and regression model fits across a FacetGrid.

This function combines regplot() and FacetGrid. It is intended as a convenient interface to fit regression models across conditional subsets of a dataset.

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

### Using a Grid

`hue`, `col`, `row` strings

Variables that define subsets of the data, which will be drawn on separate facets in the grid. 

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],palette='Set2',scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,row='smoker',scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='time',markers=['v'],scatter_kws={'s':50})
```

### Aspect and Size

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,height=10)
```

## Grids

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

### PairGrid

Subplot grid for plotting pairwise relationships in a dataset.

This object maps each variable in a dataset onto a column and row in a grid of multiple axes. Different axes-level plotting functions can be used to draw bivariate plots in the upper and lower triangles, and the marginal distribution of each variable can be shown on the diagonal.

```python
grid = sns.PairGrid(iris)
grid.map(plt.scatter)
```

Plot pairwise relationships in a dataset.

By default, this function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.

```python
sns.pairplot(iris)
```

Using different plot types from seaborn and matplotlib to for different zone of pair grid

```python
grid = sns.PairGrid(iris)
grid.map_diag(plt.hist)
grid.map_upper(plt.scatter)
grid.map_lower(sns.kdeplot)
```

### Facet Grid

Multi-plot grid for plotting conditional relationships.



```python
grid = sns.FacetGrid(data=tips, col='time',row='smoker')
```

```python
grid = sns.FacetGrid(tips, col='time',row='smoker')
grid.map(plt.hist,'total_bill')
```

```python
grid = sns.FacetGrid(tips, col='time',row='smoker',hue='sex')
grid.map(plt.scatter,'total_bill','tip')
```

### JoinGrid

Grid for drawing a bivariate plot with marginal univariate plots.

Many plots can be drawn by using the figure-level interface jointplot(). Use this class directly when you need more flexibility.

```python
grid = sns.JointGrid(x='total_bill',y='tip',data=tips)
```

```python
grid = sns.JointGrid(x='total_bill',y='tip',data=tips)
grid.plot(sns.regplot,sns.distplot)
```
