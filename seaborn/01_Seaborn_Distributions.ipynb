## Disturbutions Plots

In this notebook we focus on Disturbutions Plots in  seaborn library

### displot

Plot univariate or bivariate distributions using kernel density estimation.

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

```python
sns.kdeplot(mpg['horsepower'])
```

Figure-level interface for drawing distribution plots onto a FacetGrid.

This function provides access to several approaches for visualizing the univariate or bivariate distribution of data, including subsets of data defined by semantic mapping and faceting across multiple subplots. 

```python
sns.displot(mpg['horsepower'])
```

set `kde` to True

```python
sns.displot(mpg['horsepower'], kde=True)
```

Changing number of `bins`

```python
sns.displot(mpg['horsepower'],kde=False,bins=50)
```

### jointplot

Draw a plot of two variables with bivariate and univariate graphs.

This function provides a convenient interface to the JointGrid class, with several canned plot kinds. This is intended to be a fairly lightweight wrapper; if you need more flexibility, you should use JointGrid directly.

```python
sns.jointplot(x='horsepower',y='mpg',data=mpg,kind='scatter')
```

Changing kind to `hex`

```python
sns.jointplot(x='horsepower',y='mpg',data=mpg,kind='hex')
```

```python
mpg[['horsepower','mpg']].corr()
```

Changing kind to `reg`

```python
sns.jointplot(x='horsepower',y='mpg',data=mpg,kind='reg')
```

Changing kind to `kde`

```python
sns.jointplot(x='horsepower',y='mpg',data=mpg,kind='kde')
```

### pairplot

Plot pairwise relationships in a dataset.

By default, this function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.

It is also possible to show a subset of variables or plot different variables on the rows and columns.

```python
sns.pairplot(mpg)
```

using `hue` in pairplot

```python
sns.pairplot(mpg,hue='cylinders')
```

### rugplot

Plot marginal distributions by drawing ticks along the x and y axes.

This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

```python
sns.rugplot(mpg['horsepower'])
```

Lets using `hue` and Variable `x`,`y` that specify positions on the x and y axes.

```python
sns.rugplot(x='mpg',y='horsepower',data=mpg,hue='cylinders',palette='Set1')
```

### Multi Plots

We can plot two method in one figure as follow

```python
sns.displot(mpg['horsepower'],kde = True,color='r')
sns.rugplot(mpg['horsepower'],color='b')
```
